```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from datetime import datetime, timedelta

# API key NewsAPI
api_key = '50de3035048c4a0eb389d6ae61783df9'
url = 'https://newsapi.org/v2/everything'
# url = 'https://newsapi.org/v2/top-headlines/'
# GET https://newsapi.org/v2/top-headlines/sources?category=businessapiKey=API_KEY
# Parameter pencarian
# Tanggal hari ini dan 30 hari yang lalu
today = datetime.now().strftime('%Y-%m-%d')
last_month = (datetime.now() - timedelta(days=30)).strftime('%Y-%m-%d')
last_week = (datetime.now() - timedelta(weeks=1)).strftime('%Y-%m-%d')

# Parameter pencarian dengan rentang waktu
parameters = {
    'q': 'fufufafa',
    # 'sources':'kompas',
    'language': 'id',
    'from': last_week,
    'to': today,
    'apiKey': api_key,
    'pageSize': 20,
    'page': 1
}
```

```python

# Fungsi untuk mengambil berita
def get_news(api_url, params):
    response = requests.get(api_url, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        return None
```

```python

# Fungsi untuk scraping isi artikel
def scrape_article_content(url):
    try:
        # Mengirim request ke URL artikel
        response = requests.get(url)

        # Jika status request 200 (sukses)
        if response.status_code == 200:
            # Membuat objek BeautifulSoup untuk parsing HTML
            soup = BeautifulSoup(response.text, 'html.parser')

            # Mencari tag <p> untuk mengambil isi paragraf artikel
            paragraphs = soup.find_all('p')

            # Menggabungkan semua teks dari paragraf menjadi satu
            article_content = ' '.join([para.get_text() for para in paragraphs])

            return article_content
        else:
            print(f"Error {response.status_code}: Tidak dapat mengakses artikel {url}")
            return None
    except Exception as e:
        print(f"Terjadi kesalahan: {e}")
        return None

# Panggil API untuk mendapatkan berita
news_data = get_news(url, parameters)

# Cek jika data berhasil diambil
if news_data and news_data.get('articles'):
    articles = news_data['articles']
    print(f"Berhasil mengambil {len(articles)} artikel berita.")

    # List untuk menyimpan hasil berita
    results = []

    # Tampilkan judul berita, link, dan isi artikel
    for article in articles:
        title = article['title']
        link = article['url']
        article_content = scrape_article_content(link)

        if article_content:
            print(f"Judul: {title}")
            print(f"Link: {link}")
            print("Isi artikel berhasil di-scrape:")
            print(article_content)

            # Simpan data ke dalam list
            results.append({
                'title': title,
                'link': link,
                'content': article_content,
                'publishedAt': article['publishedAt']
            })

            print("--------------------------------------\n\n")
        else:
            print("Gagal scraping")
    
    # Simpan hasil ke dalam format CSV
    df = pd.DataFrame(results)
    df.to_csv('berita_terorisme.csv', index=False, encoding='utf-8-sig')
    print("Data berhasil disimpan ke 'berita_terorisme.csv'")
else:
    print("Gagal mengambil berita")
```
