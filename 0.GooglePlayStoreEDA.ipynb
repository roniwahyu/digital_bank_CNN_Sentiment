```python
# Assuming your data is stored in a DataFrame called 'df'
# Replace 'df' with the actual DataFrame variable name in your code

# Filter the data for entries with replies
replied_data = bank[bank['repliedAt'].notnull()]

# Group the replied data by 'appId' and count the occurrences
replied_count_by_appid = replied_data.groupby('appId').size()

# Print the count of replied entries grouped by 'appId'
print(replied_count_by_appid)
```

```python
# Assuming your data is stored in a DataFrame called 'df'
# Replace 'df' with the actual DataFrame variable name in your code

# Filter the data for entries with scores 1 to 5
filtered_data = bank[bank['score'].isin([1, 2, 3, 4, 5])]

# Create a custom aggregation function to count occurrences
def count_occurrences(arr):
    return arr.count()

# Pivot the filtered data with 'appId' as index, 'score' as columns, and 'repliedAt' as values
pivot_table = pd.pivot_table(filtered_data, index='appId', columns='score', values='repliedAt', aggfunc=count_occurrences, fill_value=0)

# Print the pivot table with counts of 'repliedAt' occurrences by 'appId' and 'score'
print(pivot_table)
```
